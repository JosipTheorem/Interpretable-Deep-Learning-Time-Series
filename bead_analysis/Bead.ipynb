```python
import glob
import os
import numpy as np
from scipy.io import loadmat

def echo_length(echo_dir):
    mats = glob.glob(os.path.join(echo_dir, "bead_*.mat"))
    max_frame = 0
    for p in mats:
        b = loadmat(p)["bsec"]
        max_frame = max(max_frame, int(np.max(b[:,2])))
    return len(mats), max_frame

base = r"C:\Users\dujme\Desktop\Seminar_venv"
for k in range(1, 13):
    d = os.path.join(base, f"10rad_60perc_echo{k}")
    n_beads, n_frames = echo_length(d)
    print(k, n_beads, n_frames)
```
